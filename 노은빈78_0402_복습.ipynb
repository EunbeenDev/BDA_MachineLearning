### KNN Algorithm, Kmeans
* KNN 지도학습(회귀, 분류), KMeans 비지도학습

### K-means
* k개를 지정하여 평균을 통해 군집화
* 비지도학습이라는 것은 (정답을 준 것이 아니라) 정답이 없는 데이터에서 머신에게 패턴을 찾아달라고 하는 것
* y라벨이 없는 상태에서 데이터를 집어넣고 특성을 군집화
* 군집화는 같은 패턴끼리 묶는다.
* 거리의 기반으로 작동(유클리디안 거리)
* k-means의 k는 클러스터(군집)의 개수
* k값은 사용자가 지정하는 값
#### 단점
* 거리기반으로만 작동하기 때문에 거리가 55, 45의 비슷한 차이여도 55에 가깝게 하드클러스팅을 한다는 점에서 문제가 된다.
* 거리 기반이기 때문에 이상치에 민감하다.
* 단순히 거리기반의 평균으로 작동하기 때문에 그룹 자체의 분산 구조는 반영하지 못한다.
* 데이터의 양이 많기 때문에 반복 횟수가 많아지면 속도가 느려질 수 있다.
* 몇 개의 군집을 선택해야 하는지 지정해 주는 것이 어렵다.
#### 장점
* 군집화 과정에서 가장 많이 활용
* 이해가 쉬움
* 간결한 알고리즘(거리 기반 이해)

### kmeans 사용 코드

```python
kmeans=KMeans(n_clusters=3,init='k-means++', max_iter=300, random_state=111)
```

```python
kmeans
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

### kmeans train
* fit 안에 넣는다.
* 기존 class column과 비교해서 kmeans class를 제외한 다른 column으로 군집화를 진행 했을 때 kmeans의 경우가 더 잘 군집화 되었는지 확인

```python
df.columns
```

```python
df_sp=df[['alcohol', 'malic_acid', 'ash', 'alcalinity_of_ash', 'magnesium',
       'total_phenols', 'flavanoids', 'nonflavanoid_phenols',
       'proanthocyanins', 'color_intensity', 'hue',
       'od280/od315_of_diluted_wines', 'proline']]

df_sp
# class를 제외한 wine data
```

```python
kmeans.fit(df_sp)
```

```python
#실제로 군집화가 어떻게 되었는지 확인
print(kmeans.labels_)
#자동으로 군집화된 내용을 보여줌
```

```python
#cluster: 비지도학습으로 만든 column
df_sp['cluster']=kmeans.labels_
```

```python
df_sp['class']=df['class']
```

```python
df_sp
```

```python
df_sp.groupby(['class','cluster'])['proline'].count()
```

### 확인 결과
* 기존 군집과 비교했을 때 결과값이 조금 다름을 알 수 있음.
* kmeans는 비지도학습이기 때문에 정답이 존재하지 않기 때문에 기존 class와 비교한 것.
* kmeans의 원리는 평균기반 거리이기 때문에 기존 class와 군집이 다르게 나올 수 있다. 

```python
df_sp.groupby(['class'])['proline'].count()
```

```python
df_sp.groupby(['cluster'])['proline'].count()
```

* 해당 column은 차원이 10개가 넘는 고차원인 상태이다.
* 고차원 상태의 data로 거리기반으로 군집화함.
* 무작정 kmeans를 모든 column에 대해 진행하면 차원의 저주라고 하는 문제가 발생할 수 있음.
* column이 많아지면 고차원인 경우, 거리가 미세하게 바뀌거나 거의 바뀌지 않을 수 있다.-> 군집화가 제대로 이루어지지 않음.

#### 어떤 기준으로 군집화할 column을 정해야 하는가? 
* data 내 유의미한  변수 사용 ->분석가가 직접 column 선별 가능
* 기술적인 측면에서는 feature에 대한 중요성을 기반으로 칼럼을 추출할 수 있다.
##### -> 모든 column들을 kmeans에 사용하는 것은 잘못된 방향이다!

#### 군집화를 통해 얻을 수 있는 insight?
* 군집이 형성된다-> 해당 군집에 따른 data의 pattern이 다를 것이다.
* 군집에 따른 다른 columns의 값의 분포
* 비지도학습이기 때문에 군집이 없는 상태였지만 비교대상으로 비교 class를 생성했었다.
* 흩어진 data 분포에서 군집화가 되었으면, 해당 군집에 대한 다른 column들의 data 특성으로 새로운 insight를 도출해낼 수 있다.
