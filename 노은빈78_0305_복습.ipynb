```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
import pandas as pd
```

```python

iris=load_iris()
```

```python
iris_data=iris.data
```

```python
iris_data
```

```python
iris_label=iris.target
```

```python
iris_label
```

```python
iris_df=pd.DataFrame(data=iris_data,columns=iris.feature_names)
iris_df['y']=iris.target
```

```python
iris_df
```

### 학습용 데이터(train data), 테스트용 데이터(test data) 분리
* 학습 데이터로 학습된 모델의 성능을 평가하기 위해 test data 필요
* -사이킷런에서 제공되는 trin_test_split() API 사용
* test(데이터 분할 가능), test_size(크기 조정 가능)
* 직접 train, test data의 데이터 비중 설정 가능 

```python
iris_df
```

```python
# 데이터 분리
X_train, X_test, y_train, y_test=train_test_split(iris_data,iris_label,test_size=0.3,random_state=111)
#test_size: 전체 데이터에서의 비중을 뜻함
#random_state: 섞는 개념
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

### DecisionTree Model
* 주어진 input 값들의 조합에 대한 의사결정규칙(rule)에 따라 출력값을 예측하는 모형. 트리구조의 그래프로 표현.
* 다른 모델들에 비해 예측력은 떨어지나 해석은 뛰어남.
* regression tree, classification tree로 나뉨.

### regression tree(회귀나무)
* 출력변수가 연속형

### classification tree(분류나무)
* 출력변수가 범주형


```python
df_clf=DecisionTreeClassifier(random_state=111)
```

### random_state
* 학습/테스트용 데이터 세트를 생성하기 위해 주어지는 난수 값
* 매번 분석할때마다 훈련/테스트셋이 변화하는걸 막기 위해 고정시키는 역할
* random_state에 특정 숫자를 입력하면, 해당 숫자를 입력하는 한 항상 같은 결과가 출력

```python
df_clf.fit(X_train, y_train)
#x_train, y_train x 데이터와 x 데이더에 대응하는 y값
```

```python
#예측 진행
pred=df_clf.predict(X_test)
```

```python
pred
#정답 출력--X_test값을 학습해 y값 에측
```

```python
#정확도 측정
from sklearn.metrics import accuracy_score
```

```python
print('DT 예측 정확도: {0:.3f}'.format(accuracy_score(y_test,pred)))
# y_test(기존의 정답)과 predO(예측값) 사이의 정확도 측정
```
